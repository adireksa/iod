```python
## VARIANCE FORMULA

def variance(data):
# Number of observations
    n = len(data)
# Mean of the data
    mean = sum(data) / n
# Square deviations
    deviations = [(x - mean) ** 2 for x in data]
# Variance
    variance = sum(deviations) / n
    return variance

variance([4, 8, 6, 5, 3, 2, 8, 9, 2, 5])
```

```python
import statistics
## variarance for population with n
pvar = statistics.pvariance([4, 8, 6, 5, 3, 2, 8, 9, 2, 5])

## variarance for population using sample set with n-1
var = statistics.variance([4, 8, 6, 5, 3, 2, 8, 9, 2, 5])
print(pvar)
print(var)

## stdev for population n
pstdev = statistics.pstdev([4, 8, 6, 5, 3, 2, 8, 9, 2, 5])
## stdev for population using sample set with n-1
stdev = statistics.stdev([4, 8, 6, 5, 3, 2, 8, 9, 2, 5])
## print
print(pstdev)
print(stdev)
```
